# Nursery Pathfinder Workflow

A command notebook for recording Mineflayer pathfinder-teacher sessions, training the action-conditioned visual world model, and inspecting recording quality. Run cells top to bottom, changing the configuration cell before launching a live run.

```python
from __future__ import annotations

import json
import os
import subprocess
import sys
from pathlib import Path

ROOT = Path.cwd()
if not (ROOT / "pyproject.toml").exists() and (ROOT.parent / "pyproject.toml").exists():
    ROOT = ROOT.parent

PYTHON = sys.executable
print(f"repo: {ROOT}")
print(f"python: {PYTHON}")
```

```python
def run(cmd, *, cwd=ROOT, env=None, check=True):
    """Run a command from the notebook and stream output."""
    merged_env = os.environ.copy()
    if env:
        merged_env.update({k: str(v) for k, v in env.items() if v is not None})
    print(">", " ".join(map(str, cmd)))
    result = subprocess.run(
        list(map(str, cmd)),
        cwd=str(cwd),
        env=merged_env,
        text=True,
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT,
    )
    print(result.stdout)
    if check and result.returncode != 0:
        raise RuntimeError(f"command failed with exit code {result.returncode}")
    return result
```

## Environment

Set these for a live server. The nursery command defaults to `remote` when `CCR_MINECRAFT_HOST` is set; otherwise it defaults to `simulated`.

```python
LIVE_ENV = {
    "CCR_MINECRAFT_HOST": "localhost",
    "CCR_MINECRAFT_PORT": "25565",
    "CCR_MINECRAFT_USERNAME": "CCRAgent",
    # "CCR_MINECRAFT_VERSION": "1.20.4",
    # "CCR_MINECRAFT_AUTH": "offline",
    # Force bridge pixel mode if needed: viewer or grid.
    "CCR_MINECRAFT_PIXELS": "viewer",
    "CCR_MINECRAFT_COMMAND_DELAY_MS": "50"
}

RECORD_DIR = ROOT / "sessions"
SMOKE_RECORD_DIR = ROOT / "sessions" / "smoke-pathfinder"
OUT_DIR = ROOT / "models" / "nursery"
SMOKE_OUT_DIR = OUT_DIR / "smoke"
REPORT = OUT_DIR / "pathfinder_report.json"

print(json.dumps(LIVE_ENV, indent=2))
```

## Checks

Verify the JavaScript bridge parses and the nursery command is visible.

```python
run(["npm.cmd", "run", "check"], cwd=ROOT / "bridge" / "mineflayer")
run([PYTHON, "-m", "cognitive_runtime", "nursery", "list"])
```

## Fast Simulated Smoke Run

This does not need Minecraft. It checks the Python path, recording loop, dataset builder, and trainer wiring.

```python
SMOKE_REPORT = OUT_DIR / "simulated_smoke_report.json"
run([
    PYTHON, "-m", "cognitive_runtime", "nursery", "run",
    "--backend", "simulated",
    "--record-dir", SMOKE_RECORD_DIR,
    "--train-episodes", "2",
    "--holdout-episodes", "1",
    "--episode-ticks", "40",
    "--world-size", "18",
    "--horizons", "1", "3",
    "--latent-width", "16",
    "--hidden-dim", "32",
    "--reconstruction-size", "8",
    "--epochs", "2",
    "--warmup-frames", "2",
    "--rollout-frames", "3",
    "--batch-size", "16",
    "--no-setup-live-arena",
    "--allow-grid-pixels",
    "--no-require-learning",
    "--out-dir", SMOKE_OUT_DIR,
    "--report", SMOKE_REPORT,
])
```

## Live Pathfinder Nursery Run

This records all train episodes, records holdout episodes, then trains offline from the recorded train sessions and evaluates on holdouts. It does not train while the Mineflayer scenario is running.

```python
run([
    PYTHON, "-m", "cognitive_runtime", "nursery", "run",
    "--backend", "remote",
    "--record-dir", RECORD_DIR,
    "--train-episodes", "6",
    "--holdout-episodes", "2",
    "--episode-ticks", "500",
    "--horizons", "1", "4", "8",
    "--out-dir", OUT_DIR,
    "--report", REPORT,
], env=LIVE_ENV)
```

## Inspect The Report

```python
report_path = REPORT if REPORT.exists() else SMOKE_REPORT
MODEL_PATH = OUT_DIR / "pathfinder-world-model.pt" if report_path == REPORT else SMOKE_OUT_DIR / "pathfinder-world-model.pt"
if report_path == SMOKE_REPORT and not REPORT.exists():
    print(f"live report not found at {REPORT}; loading smoke report instead")
    print("NOTE: smoke sessions use simulated grid/top-down pixels; run the Live Pathfinder Nursery cell for first-person exports")
report = json.loads(report_path.read_text(encoding="utf-8"))
print(f"loaded: {report_path}")
print(f"model:  {MODEL_PATH}")
print("train sessions:", *report.get("train_sessions", []), sep="\n  ")
print("holdout sessions:", *report.get("holdout_sessions", []), sep="\n  ")
print("ticks_per_frame:", report.get("ticks_per_frame"))
print("horizon_frames:", report.get("horizon_frames"))
print("learning_check:", json.dumps(report.get("learning_check", {}), indent=2))
print("metrics:", json.dumps(report.get("metrics", {}), indent=2)[:4000])
```

## Data Quality Summary

Use this after any run to see whether frames are changing, whether the bot moved, and whether the bridge reported first-person viewer pixels.

```python
from cognitive_runtime.runtime.replay import list_episodes
from cognitive_runtime.training.nursery import measure_pathfinder_recording

session_dirs = [Path(p) for p in report.get("train_sessions", []) + report.get("holdout_sessions", [])]
for session_dir in session_dirs:
    for episode_id in list_episodes(str(session_dir)):
        q = measure_pathfinder_recording(str(session_dir), episode_id)
        print(session_dir.name, episode_id)
        print(f"  frames: {q.n_frames}, unique: {q.unique_frames} ({q.unique_frame_fraction:.1%})")
        print(f"  displacement: net={q.net_displacement:.2f}, max={q.max_displacement:.2f}, blocks/tick={q.blocks_per_tick:.4f}")
        print(f"  pixel_sources: {q.pixel_sources}")
        print(f"  actions: {q.action_counts}")
```

## Export Viewer Predictions

`nursery run --out-dir` exports `predictions_episode_00000.json` automatically. Re-run this cell if you trained earlier and need to export from the saved action-world-model checkpoint.

```python
from cognitive_runtime.runtime.replay import list_episodes
from cognitive_runtime.training.action_world_model import load_action_world_model
from cognitive_runtime.training.prediction_export import export_action_session_predictions

if report_path == SMOKE_REPORT:
    print("NOTE: exporting smoke predictions; these are simulated grid/top-down pixels, not first-person")
session_dirs = [Path(p) for p in report.get("train_sessions", []) + report.get("holdout_sessions", [])]
missing = []
for session_dir in session_dirs:
    for episode_id in list_episodes(str(session_dir)):
        pred_path = session_dir / f"predictions_{episode_id}.json"
        if not pred_path.exists():
            missing.append(pred_path)

if missing:
    model, _stats = load_action_world_model(str(MODEL_PATH))
    written = export_action_session_predictions(model, [str(p) for p in session_dirs], report["horizon_frames"])
    print(f"exported {len(written)} prediction file(s)")
else:
    print("prediction files already exported")

for session_dir in session_dirs:
    for episode_id in list_episodes(str(session_dir)):
        print(session_dir.name, episode_id, (session_dir / f"predictions_{episode_id}.json").exists())
```
